```python
#EXERCISE NO.1
class Book:
    def __init__(self, title, author):
        self.title = title
        self.author = author
        self.is_available = True

class Library:
    def __init__(self):
        self.books = []

    def add_book(self, title, author):
        book = Book(title, author)
        self.books.append(book)
        print(f"Book '{title}' by {author} added successfully.")

    def show_books(self):
        print("\n--- Library Books ---")
        if not self.books:
            print("No books available.")
        else:
            for i, book in enumerate(self.books, 1):
                status = "Available" if book.is_available else "Borrowed"
                print(f"{i}. '{book.title}' by {book.author} — [{status}]")

    def borrow_book(self, title):
        for book in self.books:
            if book.title.lower() == title.lower():
                if book.is_available:
                    book.is_available = False
                    print(f"You have successfully borrowed '{book.title}'.")
                else:
                    print(f"Sorry, '{book.title}' is already borrowed.")
                return
        print(f"Book '{title}' does not exist in the library.")

    def return_book(self, title):
        for book in self.books:
            if book.title.lower() == title.lower():
                if not book.is_available:
                    book.is_available = True
                    print(f"You have successfully returned '{book.title}'.")
                else:
                    print(f"'{book.title}' was not borrowed.")
                return
        print(f"Book '{title}' does not exist in the library.")


# --- Main Program ---
library = Library()

# Add books
num_books = int(input("How many books do you want to add? "))
for _ in range(num_books):
    title = input("Enter book title: ")
    author = input("Enter book author: ")
    library.add_book(title, author)

while True:
    print("\n--- Library Menu ---")
    print("1. Show all books")
    print("2. Borrow a book")
    print("3. Return a book")
    print("4. Exit")

    choice = input("Enter your choice: ")

    if choice == "1":
        library.show_books()
    elif choice == "2":
        title = input("Enter the title of the book to borrow: ")
        library.borrow_book(title)
    elif choice == "3":
        title = input("Enter the title of the book to return: ")
        library.return_book(title)
    elif choice == "4":
        print("Goodbye!")
        break
    else:
        print("Invalid choice. Please try again.")
```

```python
# EXERCISE NO.2
class Employee:
    def __init__(self, name, hourly_wage, hours_worked):
        if hourly_wage < 0 or hours_worked < 0:
            raise ValueError("Hourly wage and hours worked cannot be negative.")
        self.name = name
        self.hourly_wage = hourly_wage
        self.hours_worked = hours_worked

    def compute_salary(self):
        if self.hours_worked <= 160:
            base_salary = self.hourly_wage * self.hours_worked
            overtime_pay = 0
        else:
            base_salary = self.hourly_wage * 160
            overtime_hours = self.hours_worked - 160
            overtime_pay = overtime_hours * self.hourly_wage * 1.5
        return base_salary, overtime_pay

    def display_payroll(self):
        base_salary, overtime_pay = self.compute_salary()
        total = base_salary + overtime_pay
        print(f"\n--- Payroll for {self.name} ---")
        print(f"Hours Worked   : {self.hours_worked} hrs")
        print(f"Hourly Wage    : ${self.hourly_wage:.2f}")
        print(f"Base Salary    : ${base_salary:.2f}")
        if overtime_pay > 0:
            print(f"Overtime Pay   : ${overtime_pay:.2f}")
        print(f"Total Salary   : ${total:.2f}")


# --- Main Program ---
employees = []

while True:
    try:
        num_employees = int(input("Enter number of employees: "))
        if num_employees < 1:
            print("Please enter at least one employee.")
        else:
            break
    except ValueError:
        print("Invalid input. Please enter a valid number.")

for _ in range(num_employees):
    name = input("\nEnter employee name: ")
    while True:
        try:
            hourly_wage = float(input(f"Enter hourly wage for {name}: $"))
            hours_worked = float(input(f"Enter total hours worked for {name}: "))
            emp = Employee(name, hourly_wage, hours_worked)
            employees.append(emp)
            break
        except ValueError as e:
            print(f"Error: {e}. Please try again.")

print("\n========== PAYROLL SUMMARY ==========")
for emp in employees:
    emp.display_payroll()
```

```python
class Product:
    def __init__(self, name, price, stock):
        self.name = name
        self.price = price
        self.stock = stock

class ShoppingCart:
    def __init__(self):
        self.cart = []

    def show_products(self, products):
        print("\n--- Available Products ---")
        for i, product in enumerate(products, 1):
            print(f"{i}. {product.name} — ${product.price:.2f} (Stock: {product.stock})")

    def add_to_cart(self, product, quantity):
        if quantity > product.stock:
            print(f"Sorry, only {product.stock} unit(s) of '{product.name}' available.")
            return
        if quantity <= 0:
            print("Invalid quantity. Please enter a positive number.")
            return
        product.stock -= quantity
        for item in self.cart:
            if item["product"].name == product.name:
                item["quantity"] += quantity
                print(f"Added {quantity} more '{product.name}' to cart.")
                return
        self.cart.append({"product": product, "quantity": quantity})
        print(f"'{product.name}' (x{quantity}) added to cart.")

    def remove_from_cart(self, product_name):
        for item in self.cart:
            if item["product"].name.lower() == product_name.lower():
                item["product"].stock += item["quantity"]
                self.cart.remove(item)
                print(f"'{product_name}' removed from cart.")
                return
        print(f"'{product_name}' is not in your cart.")

    def show_cart(self):
        print("\n--- Your Cart ---")
        if not self.cart:
            print("Your cart is empty.")
            return
        total = 0
        for item in self.cart:
            subtotal = item["product"].price * item["quantity"]
            total += subtotal
            print(f"  {item['product'].name} x{item['quantity']} — ${subtotal:.2f}")
        print(f"Total: ${total:.2f}")

    def checkout(self):
        if not self.cart:
            print("Your cart is empty. Nothing to checkout.")
            return
        self.show_cart()
        print("Thank you for your purchase! Goodbye!")
        self.cart.clear()


# --- Predefined Products ---
products = [
    Product("Laptop", 999.99, 5),
    Product("Mouse", 29.99, 10),
    Product("Keyboard", 49.99, 8),
    Product("Monitor", 299.99, 3),
    Product("Headphones", 79.99, 6)
]

cart = ShoppingCart()

while True:
    print("\n--- Shopping Menu ---")
    print("1. View products")
    print("2. Add product to cart")
    print("3. Remove product from cart")
    print("4. View cart")
    print("5. Checkout")
    print("6. Exit")

    choice = input("Enter your choice: ")

    if choice == "1":
        cart.show_products(products)

    elif choice == "2":
        cart.show_products(products)
        try:
            product_choice = int(input("Enter product number to add: ")) - 1
            if 0 <= product_choice < len(products):
                quantity = int(input("Enter quantity: "))
                cart.add_to_cart(products[product_choice], quantity)
            else:
                print("Invalid product selection.")
        except ValueError:
            print("Invalid input. Please enter a number.")

    elif choice == "3":
        cart.show_cart()
        product_name = input("Enter product name to remove: ")
        cart.remove_from_cart(product_name)

    elif choice == "4":
        cart.show_cart()

    elif choice == "5":
        cart.checkout()
        break

    elif choice == "6":
        print("Goodbye!")
        break

    else:
        print("Invalid choice. Please try again.")
```
